## Working with 'Orders' workbook Excel sheet - Toys

```python
#Importing the necessary dependencies
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the orders excel sheet
orders_df = pd.read_excel('order_list.xlsx',sheet_name = 'Toys',header=1)
```

```python
#Viewing the first 10 rows of the orders excel sheet
orders_df.head(10)
```

```python
orders_df.describe()
```

```python
toys_order_city_list = list(orders_df['City'].unique())
```

```python
len(toys_order_city_list)
```

```python
toys_order_state_list = list(orders_df['state'].unique())
```

```python
len(toys_order_state_list)
```

```python
# Dropping the irrelevant columns from data set which will not be used for doing the analysis
orders_df = orders_df.drop(['Address 1','Adress 2','Adress 3'], axis = 1)
```

```python
orders_df
```

```python
toys_state_name = list(orders_df.state.value_counts().keys())
```

```python
toys_state_name
```

```python
toys_state_count = orders_df.state.value_counts().tolist()
```

```python
toys_state_count
```

```python
toys_state_count_df = pd.DataFrame(list(zip(toys_state_name, toys_state_count)),
               columns =['State_Name', 'order_count'])
```

```python
toys_state_count_df
```

```python
toys_plot_df = toys_state_count_df[toys_state_count_df['order_count'] >= 75]
```

```python
toys_plot_df
```

### Plotting only those states who have more than 75 orders for toys section from Yoshops

```python
x = list(toys_plot_df['State_Name'])
y = list(toys_plot_df['order_count'])
```

```python
fig = plt.figure(figsize=(14, 20))
plt.barh(x, y)
plt.ylabel('State Name')
plt.xlabel('Order Count')
for index, value in enumerate(y):
    plt.text(value, index,
             str(value))
plt.title('Orders from each state having more than 75 orders')
plt.show()
```
