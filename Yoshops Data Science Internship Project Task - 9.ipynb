```python
from PIL import Image, ImageDraw, ImageFont

# Banner
banner_text = "Data Science Basic Training Program"
banner_color = (0, 128, 255)  # RGB color (blue)
banner_size = (800, 200)  # Width, Height in pixels
font_path = "arial.ttf"  # Replace with the path to your desired font file

# Tagline
tagline = "Unlock the Power of Data: Join Our Basic Training Program for All Ages"

# Topic Description
title = "Data Science Basic Training Program for Everyone (Age 10 to 60)"
description = "Our comprehensive data science basic training program is designed for individuals of all ages, from 10 to 60 years old. Whether you're a student, professional, or retiree, this program will equip you with the fundamental knowledge and skills needed to embark on a successful data science journey. Join us and discover the fascinating world of data science as we cover key concepts, techniques, and applications that are essential in today's data-driven world. No prior experience required!"

# Keywords
keywords = [
    "Data science",
    "Training program",
    "Basic",
    "All ages",
    "10 to 60",
    "Student",
    "Professional",
    "Retiree",
    "Fundamental knowledge",
    "Skills",
    "Data-driven",
    "Concepts",
    "Techniques",
    "Applications",
    "No prior experience"
]

# Create banner image
banner_image = Image.new("RGB", banner_size, color=banner_color)
draw = ImageDraw.Draw(banner_image)
font = ImageFont.truetype(font_path, size=48)
text_width, text_height = draw.textsize(banner_text, font=font)
text_position = ((banner_size[0] - text_width) // 2, (banner_size[1] - text_height) // 2)
draw.text(text_position, banner_text, fill="white", font=font)

# Save banner image
banner_image.save("banner_2.jpg")

# Write tagline to a text file
with open("tagline.txt", "w") as file:
    file.write(tagline)

# Write topic description to a text file
with open("topic_description.txt", "w") as file:
    file.write(f"Title: {title}\n")
    file.write(f"Description: {description}")

# Write keywords to a text file
with open("keywords.txt", "w") as file:
    file.write("\n".join(keywords))
```

### B. Create 200sec ðŸŽ¥ shorts videos(MP4) on Data Science Basic Training Program for Everyone.

```python
video_url = "https://yoshops.com/products?keywords=data+science"
video_uClient = uReq(video_url)
video_page_html = video_uClient.read()
video_uClient.close()
video_page_soup = soup(video_page_html,"html.parser")

video_containers = video_page_soup.findAll("div", {"class":"col-sm-3 col-xs-6"})

video_container =  video_containers[0]
image1_link_video = video_container.img['src']
im1 = Image.open(requests.get(image1_link_video, stream=True).raw)
im1.save(r"C:\Users\Advait\Desktop\Yoshops Internships\Project Task - 9\videoimages\im3.webp")

video_container =  video_containers[1]
image2_link_video = video_container.img['src']
im2 = Image.open(requests.get(image2_link_video, stream=True).raw)
im2.save(r"C:\Users\Advait\Desktop\Yoshops Internships\Project Task - 9\videoimages\im4.webp")


#Making video from images
width = 1280
height = 720
channel = 3
fps = 1
sec = 200
fourcc = cv2.VideoWriter_fourcc(*'MP42')
video = cv2.VideoWriter('video.mp4', fourcc, float(fps), (width, height))

directory = r'C:\Users\Advait\Desktop\Yoshops Internships\Project Task - 9\videoimages'
img_name_list = os.listdir(directory)
for frame_count in range(fps*sec):
    img_name = np.random.choice(img_name_list)
    img_path = os.path.join(directory, img_name)
    img = cv2.imread(img_path)
    img_resize = cv2.resize(img, (width, height))
    video.write(img_resize)
video.release()
```
