```python
#Importing the necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Analysis for Myntra

```python
# reading the .csv file
myntra_df = pd.read_csv('HRX Store - Buy HRX Clothing _ Accessories Online in India _ MYNTRA.csv')
```

```python
myntra_df.head(10)
```

```python
# Replace the format '|\nnumber' with 'number' in a productratingscount column
myntra_df['productratingscount'] = myntra_df['productratingscount'].str.replace(r'\|\n', '')
```

```python
myntra_df.head(10)
```

```python
sns.set(style="darkgrid")  # Set the style of the plot

# Count the occurrences of each unique value in 'productratingscontainer'
rating_counts = myntra_df['productratingscontainer'].value_counts()

plt.figure(figsize=(10, 6))  # Set the figure size
sns.barplot(x=rating_counts.index, y=rating_counts.values, color='skyblue')  # Create the bar plot
plt.title('Product Ratings Distribution')  # Set the title
plt.xlabel('Rating')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.xticks(rotation=45)  # Rotate the x-axis labels for better visibility
plt.show()  # Show the plot
```

```python
sns.set(style="darkgrid")  # Set the style of the plot

# Count the occurrences of each unique value in 'Price'
price_counts = myntra_df['Price'].value_counts()

plt.figure(figsize=(20, 10))  # Set the figure size
sns.barplot(x=price_counts.index, y=price_counts.values, color='skyblue')  # Create the bar plot
plt.title('Price distribution')  # Set the title
plt.xlabel('Price')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.xticks(rotation=45)  # Rotate the x-axis labels for better visibility
plt.show()  # Show the plot
```

```python
# Highest Rating Product
highest_rating = myntra_df.groupby('productproduct')['productratingscontainer'].max()
```

```python
highest_rating.to_csv('Highest_rating_products.csv', index=True)
```

```python
highest_rating.head(15)
```

```python
# cost grouping with product name
product_name_group_cost = myntra_df.groupby('productproduct')['Price'].max()
```

```python
product_name_group_cost.head(15)
```

```python
product_name_group_cost.to_csv('Product_name_grouped_with_cost.csv', index=True)
```

### Analysis For Flipkart

```python
flipkart_df = pd.read_csv('Hrx By Hrithik Roshan Clothing And Accessories - Buy Hrx By Hrithik Roshan Clothing And Accessories .csv')
```

```python
flipkart_df.head(15)
```

```python
columns_mapping = {"irpwta":"Name of Product", "_30jeq3":"Price of Product"}
flipkart_df = flipkart_df.rename(columns=columns_mapping)
```

```python
flipkart_df.head(20)
```

```python
summary_stats = flipkart_df['Price of Product'].describe()
print(summary_stats)
```

```python
plt.figure(figsize=(10, 10))
plt.hist(flipkart_df['Price of Product'], bins=10, edgecolor='black')
plt.title('Distribution of Product Prices for HRX')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
product_counts = flipkart_df['Name of Product'].value_counts()

plt.figure(figsize=(20, 10))
sns.barplot(x=product_counts.index, y=product_counts.values, color='skyblue')
plt.title('Product Counts for HRX')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
plt.scatter(flipkart_df['Price of Product'], flipkart_df['Name of Product'], alpha=0.5)
plt.title('Product Price vs. Name for HRX')
plt.xlabel('Price')
plt.ylabel('Product')
plt.xticks(rotation=90)
plt.show()
```
