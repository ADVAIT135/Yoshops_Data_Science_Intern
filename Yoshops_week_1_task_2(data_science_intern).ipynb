#### Importing the 'openpyxl' module using 'load_workbook' function

```python
from openpyxl import load_workbook as opxl
```

```python
excel_sheet = opxl(filename='covid_19_diagnosis.xlsx')
```

```python
excel_sheet.sheetnames
```

```python
sheet = excel_sheet.active
```

```python
for value in sheet.iter_rows(max_row=5, min_col=1, max_col=5, values_only = True):
    print(value)
```
