### 9 --> Number of Orders Across Parts of a Day

```python
orders['Part_of_Day'] = pd.cut(orders['Order_Date_and_Time_Stamp'].dt.hour, bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'], include_lowest=True)

orders_by_part_of_day = orders.groupby('Part_of_Day')['Order #'].count().to_frame()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
a = orders_by_part_of_day['Order #'].index
b = orders_by_part_of_day['Order #'].values
ax.bar(a,b)
ax.set_ylabel('Number of Orders')
ax.set_xlabel('Part of Day')
ax.set_title('Number of Orders Across Parts of a Day')
for i, (x, y) in enumerate(zip(a, b)):
    plt.text(x, y, str(y), ha='center', va='bottom', fontweight='bold', fontsize=10)
plt.show()
```
