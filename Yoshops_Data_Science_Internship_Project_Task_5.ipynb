```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import csv
import requests
from bs4 import BeautifulSoup

# Open the CSV file
with open('reliance_digital_mobiles_under_10000_link.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Skip the header row if needed
    next(csv_reader)
    
    # Iterate over each row in the CSV file
    for i, row in enumerate(csv_reader, 1):  # Start the serial number from 1
        # Get the link from the appropriate column
        link = row[0]  # Adjust the column index as per your CSV structure
        
        # Make a request to the link
        response = requests.get(link)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Extract the data you need from the parsed HTML using BeautifulSoup
            # Example: Extract the title from the page
            title = soup.title.text
            
            # Do further processing with the extracted data as needed
            print(f"Item Number {i}: {title}")
        else:
            # Handle any errors or unsuccessful requests
            print(f"Error accessing link in row {i}: {link}")
```

```python
reliance_mobiles_under_10k = pd.read_csv('Reliance_Digital_Mobile_phones_under_10000_brand.csv')
```

```python
reliance_mobiles_under_10k
```

```python
# Read the CSV file into a DataFrame
df = reliance_mobiles_under_10k

# Analyze the data
#Counting the number of occurrences for each brand
brand_counts = df['Product_brand_name'].value_counts()
print("Brand Counts:")
print(brand_counts)
```

```python
# Calculating the average price by brand
average_prices = df.groupby('Product_brand_name')['Product_Price'].mean()
print("Average Prices by Brand:")
print(average_prices)


```

```python
# Find the most expensive product
max_price = df['Product_Price'].max()
most_expensive_product = df.loc[df['Product_Price'] == max_price, 'Product_title'].iloc[0]
print("Most Expensive Product:")
print(most_expensive_product)
```

```python
# Filter products based on a condition
filtered_products = df.loc[df['Product_Price'] > 5000]  # Filter products with price > 5000
print("Filtered Products:")
print(filtered_products)
```

```python
# Bar plot of brand counts
brand_counts = df['Product_brand_name'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.title('Brand Counts')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of average prices by brand
average_prices = df.groupby('Product_brand_name')['Product_Price'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=average_prices.index, y=average_prices.values)
plt.title('Average Prices by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of brand names and prices
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Product_brand_name', y='Product_Price')
plt.title('Brand Names vs Prices')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
