```python
Mobiles_price_df = Mobiles
```

```python
Mobiles_price_df[['Price','Offer']] = Mobiles_price_df['price_details'].str.split('\n', expand=True)
```

```python
Mobiles_price_df['Offer'] = Mobiles_price_df['Offer'].str[:-7]
```

```python
Mobiles_price_df
```

```python
Mobiles_price_df.drop('price_details', axis=1, inplace=True)
```

```python
Mobiles_price_df.drop('Offer', axis=1, inplace=True)
```

```python
Mobiles_price_df
Mobiles_price_df['Price'] = Mobiles_price_df['Price'].str.replace('₹', '')
```

```python
Mobiles_price_df['Price'] = Mobiles_price_df['Price'].str.replace(',', '')
```

```python
Mobiles_price_df
```

```python
Mobiles_price_df['Price'] = Mobiles_price_df['Price'].astype(str)
Mobiles_price_df = Mobiles_price_df[~Mobiles_price_df['Price'].str.contains('off')]
```
